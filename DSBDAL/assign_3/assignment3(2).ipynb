```python
junior = df1[df1['Age']>=19]
```

```python
junior = junior[junior['Age']<25]
junior
```

```python
middle = df1[df1['Age']>=25]
```

```python
middle = middle[middle['Age']<30]
middle
```

```python
senior = df1[df1['Age']>=30]
```

```python
senior
```

```python
# Statistics for junior grp
jr_mean_sal = junior['Salary'].mean()
jr_median_sal = junior['Salary'].median()
jr_max_sal = junior['Salary'].max()
jr_min_sal = junior['Salary'].min()
jr_std_sal = junior['Salary'].std()
print('Mean salary of Junior group ',jr_mean_sal)
print('Median salary of Junior group ',jr_median_sal)
print('Maximum salary of Junior group ',jr_max_sal)
print('Minimum salary of Junior group ',jr_min_sal)
print('Standard deviation of salary of Junior group ',jr_max_sal)
```

```python
# Statistics for middle grp
mid_mean_sal = middle['Salary'].mean()
mid_median_sal = middle['Salary'].median()
mid_max_sal = middle['Salary'].max()
mid_min_sal = middle['Salary'].min()
mid_std_sal = middle['Salary'].std()
print('Mean salary of Middle group ',mid_mean_sal)
print('Median salary of Middle group ',mid_median_sal)
print('Maximum salary of Middle group ',mid_max_sal)
print('Minimum salary of Middle group ',mid_min_sal)
print('Standard deviation of salary of Middle group ',mid_max_sal)
```

```python
# Statistics for senior grp
sr_mean_sal = senior['Salary'].mean()
sr_median_sal = senior['Salary'].median()
sr_max_sal = senior['Salary'].max()
sr_min_sal = senior['Salary'].min()
sr_std_sal = senior['Salary'].std()
print('Mean salary of Senior group ',sr_mean_sal)
print('Median salary of Senior group ',sr_median_sal)
print('Maximum salary of Senior group ',sr_max_sal)
print('Minimum salary of Senior group ',sr_min_sal)
print('Standard deviation of salary of Senior group ',sr_max_sal)
```
