### Importing libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

### Dataset Loading ,observation and preprocessing

```python
df=pd.read_csv("titanic.csv")
df
```

```python
df.isnull().sum()
```

```python
df['Age'].value_counts()
```

```python
df['Age'].mode()
```

```python
df['Embarked'].value_counts()
```

```python
df["Age"].fillna(df["Age"].mean(),inplace=True)
df["Cabin"].fillna(0, inplace=True)
df["Embarked"].fillna("S", inplace=True)
df
```

```python
df.isnull().sum()
```

### Heatmap

```python
df_map=df.drop(columns=["PassengerId","Name","Ticket","Cabin","Embarked"])
df_map["Sex"].replace(["male","female"],[0,1],inplace=True)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df_map.corr(),annot=True)
```
